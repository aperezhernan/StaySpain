Disponibilitat mitjana dels allotjaments 
tur√≠stics en els diferents terminis (30, 60, 90 i 365 dies) a cada ciutat? 

```python
import pandas as pd

df = pd.read_pickle(r"data\staySpain_cleaned.pkl")

columns = ["availability_30", "availability_60", "availability_90", "availability_365"]

avg_availability = (
    df.groupby("city", dropna=False)[columns]
      .mean(numeric_only=True)
      .round(1)
      .sort_index()
)
avg_availability
```

Ciudadades con mayor y menos disponiblidad media en cada periodo

```python
resultados = {}

for periodo in avg_availability.columns:
    max_valor = avg_availability[periodo].max()
    min_valor = avg_availability[periodo].min()

    ciudades_max = avg_availability[avg_availability[periodo] == max_valor].index.tolist()
    ciudades_min = avg_availability[avg_availability[periodo] == min_valor].index.tolist()

    resultados[periodo] = {
        "max_valor": max_valor,
        "ciudades_max": ciudades_max,
        "min_valor": min_valor,
        "ciudades_min": ciudades_min
    }

for periodo, info in resultados.items():
    print(f"\nPeriodo: {periodo}")
    print(f"  üîº M√°xima disponibilidad ({info['max_valor']} d√≠as): {', '.join(info['ciudades_max'])}")
    print(f"  üîΩ M√≠nima disponibilidad ({info['min_valor']} d√≠as): {', '.join(info['ciudades_min'])}")
```

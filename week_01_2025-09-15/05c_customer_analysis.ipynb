# Qué porcentaje de alojamientos tienen una valoración más alta por ciudad?

```python
# Filtramos para excluir los posibles NULL en "review_scores_rating"
filtered_df = df[df['review_scores_rating'].notnull()].copy()

filtered_df.head()
```

```python
# Creamos una nueva columna para identificar a los alojamientos con un rating superior a 80. Pasarlo a int nos convierte True a 1, False a 0
filtered_df['above_80'] = (filtered_df['review_scores_rating']  > 80).astype(int)

```

```python
# Agrupamos por "city" y calculamos las métricas
df_grouped = (
        filtered_df
        .groupby('city')
        .agg(
            total_apartments=('apartment_id', 'count'),
            apartments_above_80=('above_80', 'sum')
        )
        .assign(
            percentage_above_80=lambda x: ((x['apartments_above_80'] / x['total_apartments']) * 100).round(2)
        )
        .reset_index()
    )
```

```python
result = df_grouped.sort_values(by="percentage_above_80", ascending=False).reset_index(drop=True)

result
```
